```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_selection import RFECV
```

```python
X_train = pd.read_csv("../01_Data/X_train.csv")
X_test = pd.read_csv("../01_Data/X_test.csv")
y_train = pd.read_csv("../01_Data/y_train.csv")
y_test = pd.read_csv("../01_Data/y_test.csv")

# Convertir y_train y y_test a vectores 1D
y_train = y_train.squeeze()
y_test = y_test.squeeze()

# Variables seleccionadas
numerical_vars = [
    'P220_5', 'P220_9', 'P220_2', 'P220_3', 'P220_1_CANT_2', 'P219_UM_COD', 'P220_2_VAL',  
    'P220_1_PREC_2', 'FACTOR', 'P220_2_PRE_KG', 'P220_4', 'P219_CANT_2',
    'P220_2_PREC_1', 'P209_ANIO', 'P220_1_PRE_KG', 'P220_8', 
    'P217_SUP_ha', 'P218', 'P217_SUP_1', 'P220_7', 'P217_SUP_2', 'P220_6', 'P220_10', 
    'duracion_cosecha_meses', 'produccion_area_ratio', 'precio_valor_ratio'
]

categorical_vars = [
    'REGION', 'P204_NOM', 'P204_TIPO', 'P208', 'P212', 
    'P222_1', 'P222_2', 'P222_6', 'P223_2', 'P223_3', 'P223_5'
]


# Preprocesamiento
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_vars),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_vars)
    ]
)

# Modelo base: Regresión logística
logistic_model = LogisticRegression(max_iter=1000, random_state=42)

# RFECV para Selección de Características
rfecv = RFECV(
    estimator=logistic_model,
    step=1,  # Eliminación de una característica por iteración
    cv=5,    # Validación cruzada de 5 folds
    scoring='accuracy',  # Métrica de evaluación
    n_jobs=-1
)

# Pipeline con preprocesamiento y RFECV
rfecv_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', rfecv),
    ('classifier', logistic_model)
])

# Entrenar RFECV
rfecv_pipeline.fit(X_train, y_train)

# Predicciones
y_pred = rfecv_pipeline.predict(X_test)

# Evaluación del modelo
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Visualización de la validación cruzada usando cv_results_
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(rfecv.cv_results_['mean_test_score']) + 1), rfecv.cv_results_['mean_test_score'])
plt.xlabel("Número de características seleccionadas")
plt.ylabel("Accuracy en validación cruzada")
plt.title("RFECV: Selección de características")
plt.show()

# Variables seleccionadas
selected_features = (
    numerical_vars + 
    list(rfecv_pipeline.named_steps['preprocessor'].named_transformers_['cat'].get_feature_names_out(categorical_vars))
)

# Filtrar características seleccionadas
selected_features_final = [feature for feature, selected in zip(selected_features, rfecv.support_) if selected]

print("Características seleccionadas por RFECV:")
print(selected_features_final)

# Importancia de características
coefficients = rfecv.estimator_.coef_[0]
importance_df = pd.DataFrame({
    'Feature': selected_features_final,
    'Coefficient': coefficients
}).sort_values(by='Coefficient', key=abs, ascending=False)

print("Importancia de las características seleccionadas:")
print(importance_df)

# Visualizar las importancias
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Coefficient'])
plt.xlabel("Coeficiente")
plt.title("Importancia de las Características Seleccionadas")
plt.gca().invert_yaxis()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

# Cargar conjuntos de datos
X_train = pd.read_csv("../01_Data/X_train.csv")
X_test = pd.read_csv("../01_Data/X_test.csv")
y_train = pd.read_csv("../01_Data/y_train.csv").squeeze()
y_test = pd.read_csv("../01_Data/y_test.csv").squeeze()

# Variables seleccionadas
numerical_vars = [
    'P220_5', 'P220_9', 'P220_2', 'P220_3', 'P220_1_CANT_2', 'P219_UM_COD', 'P220_2_VAL',  
    'P220_1_PREC_2', 'FACTOR', 'P220_2_PRE_KG', 'P220_4', 'P219_CANT_2',
    'P220_2_PREC_1', 'P209_ANIO', 'P220_1_PRE_KG', 'P220_8', 
    'P217_SUP_ha', 'P218', 'P217_SUP_1', 'P220_7', 'P217_SUP_2', 'P220_6', 'P220_10', 
    'duracion_cosecha_meses', 'produccion_area_ratio', 'precio_valor_ratio'
]

categorical_vars = [
    'REGION', 'P204_NOM', 'P204_TIPO', 'P208', 'P212', 
    'P222_1', 'P222_2', 'P222_6', 'P223_2', 'P223_3', 'P223_5'
]

# Preprocesamiento
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_vars),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_vars)
    ]
)

# Pipeline con preprocesamiento y XGBoost
xgb_model = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', xgb_model)
])

# Entrenar el modelo
pipeline.fit(X_train, y_train)

# Predicciones
y_pred = pipeline.predict(X_test)
y_proba = pipeline.predict_proba(X_test)[:, 1]

# Evaluación del modelo
print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, label=f'Curva ROC (AUC = {roc_auc_score(y_test, y_proba):.4f})')
plt.plot([0, 1], [0, 1], 'k--', label='Aleatorio')
plt.title('Curva ROC - XGBoost')
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Importancia de características
xgb_model = pipeline.named_steps['classifier']
feature_names = (
    numerical_vars + 
    list(pipeline.named_steps['preprocessor'].named_transformers_['cat'].get_feature_names_out(categorical_vars))
)

importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': xgb_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

print("Importancia de las características:")
print(importance_df)

# Visualizar las importancias
plt.figure(figsize=(10, 8))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.xlabel("Importancia")
plt.title("Importancia de las Características - XGBoost")
plt.gca().invert_yaxis()
plt.show()
```
